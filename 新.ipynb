```python
#引用的库：
import pandas as pd
import numpy as np
import scipy.stats as stats 
from scipy.stats import chi2_contingency
import Tkinter as tk

#读取乳腺癌数据集（excel格式）：
fec =pd.read_excel('project.xlsx')

#计算乳腺癌复发率与患者身体各状况的相关性：
class relation(object):
    def __init__(self,list_count):
        self.list_count=list_count
    def age(self):#乳腺癌复发率与患者年龄的相关性
        list1 = np.array(fec.age)

        grouped = fec.groupby(['Class',list1])
        immie = grouped.size().unstack(0)
        print immie

        normed_sums = immie.div(immie.sum(axis=1),axis = 0)
        print normed_sums
        print normed_sums[:6].plot(kind='barh', stacked = True)#创建柱状图

        list_age = normed_sums.no_recurrence_events.tolist()
        
        r_age, p_age=stats.pearsonr(self.list_count,list_age)  
        print "患者年龄与不复发的相关系数为",r_age
        print "患者年龄与不复发的显著性值为",p_age
    def tumor(self):#乳腺癌复发率与患者肿瘤大小的相关性
        list2 = np.array(fec.tumor_size)
        grouped_tumor = fec.groupby(['Class',list2])
        immie_tumor = grouped_tumor.size().unstack(0)
        print immie_tumor

        normed_tumor = immie_tumor.div(immie_tumor.sum(axis=1),axis = 0)
        print normed_tumor
        print normed_tumor[:11].plot(kind='barh', stacked = True)#创建柱状图

        list_tumor = normed_tumor.no_recurrence_events.tolist()
        
        r_tumor, p_tumor=stats.pearsonr(self.list_count,list_tumor)  
        print "患者肿瘤大小与不复发的相关系数为",r_tumor
        print "患者肿瘤大小与不复发的显著性值为",p_tumor
    def irradiat(self):#乳腺癌复发率与患者化疗情况的相关性
        list3 = np.array(fec.irradiat)
        grouped_ra = fec.groupby(['Class',list3])
        immie_ra = grouped_ra.size().unstack(0)
        print immie_ra

        normed_ra = immie_ra.div(immie_ra.sum(axis=1),axis = 0)
        print normed_ra
        print normed_ra[:11].plot(kind='barh', stacked = True)#创建柱状图

        list_ra = normed_ra.no_recurrence_events.tolist()
        
        r_ra, p_ra=stats.pearsonr(self.list_count,list_ra)  
        print "患者经历化疗与不复发的相关系数为",r_ra
        print "患者经历化疗与不复发的显著性值为",p_ra
    def breast(self):#乳腺癌复发率与患者肿瘤位置的相关性
        list4 = np.array(fec.breast)
        grouped_breast = fec.groupby(['Class',list4])
        immie_breast = grouped_breast.size().unstack(0)
        print immie_breast

        normed_breast = immie_breast.div(immie_breast.sum(axis=1),axis = 0)
        print normed_breast
        print normed_breast[:11].plot(kind='barh', stacked = True)#创建柱状图

        list_breast = normed_breast.no_recurrence_events.tolist()
        
        r_breast, p_breast=stats.pearsonr(self.list_count,list_breast)  
        print "患者肿瘤位置与不复发的相关系数为",r_breast
        print "患者肿瘤位置与不复发的显著性值为",p_breast
    def deg(self):#乳腺癌复发率与患者恶性肿瘤程度的相关性：
        list5 = np.array(fec.deg_malig)
        grouped_deg = fec.groupby(['Class',list5])
        immie_deg = grouped_deg.size().unstack(0)
        print immie_deg

        normed_deg = immie_deg.div(immie_deg.sum(axis=1),axis = 0)
        print normed_deg
        print normed_deg[:11].plot(kind='barh', stacked = True)#创建柱状图

        list_deg = normed_deg.no_recurrence_events.tolist()
        
        r_deg, p_deg=stats.pearsonr(self.list_count,list_deg)  
        print "患者恶性肿瘤程度与不复发的相关系数为",r_deg
        print "患者恶性肿瘤程度与不复发的显著性值为",p_deg
    def inv(self):#乳腺癌复发率与患者受侵淋巴结数的相关性：
        list6 = np.array(fec.inv_nodes)
        grouped_inv = fec.groupby(['Class',list6])
        immie_inv = grouped_inv.size().unstack(0)
        print immie_inv

        normed_inv = immie_inv.div(immie_inv.sum(axis=1),axis = 0)
        print normed_inv
        print normed_inv[:7].plot(kind='barh', stacked = True)#创建柱状图

        list_inv = normed_inv.no_recurrence_events.tolist()
        
        r_inv, p_inv=stats.pearsonr(self.list_count,list_inv)  
        print "患者受侵淋巴结数与不复发的相关系数为",r_inv
        print "患者受侵淋巴结数与不复发的显著性值为",p_inv
    def quad(self):#乳腺癌复发率与患者肿瘤所在象限的相关性：
        list7 = np.array(fec.breast_quad)
        grouped_quad = fec.groupby(['Class',list7])
        immie_quad = grouped_quad.size().unstack(0)
        print immie_quad

        normed_quad = immie_quad.div(immie_quad.sum(axis=1),axis = 0)
        print normed_quad
        print normed_quad[:11].plot(kind='barh', stacked = True)#创建柱状图
        
        list_quad = normed_quad.no_recurrence_events.tolist()
        
        r_quad, p_quad=stats.pearsonr(self.list_count,list_quad)  
        print "患者肿瘤所在象限与不复发的相关系数为",r_quad
        print "患者肿瘤所在象限与不复发的显著性值为",p_quad

#创建二联表并计算卡方值：
class crosstab(object):
    def __init__(self,m,n):
        self.m=m
        self.n=n
    def tab(self):
        h = pd.crosstab(self.m,self.n,margins=True)#创建交叉表
        list_non = list(h.no_recurrence_events)
        list_re_ = list(h.recurrence_events)
        d = np.array([list_non,list_re_])
        print "卡方值为",chi2_contingency(d)[0],"P值为",chi2_contingency(d)[1],"自由度为",chi2_contingency(d)[2],"与原数据数组同维度的对应理论值为",chi2_contingency(d)[3]

#根据用户提供数据预测复发概率
class predict(object):
    def __init__(self,im1,p1,im2,p2,im3,p3):#im为权重，p为概率（以频率代替）
        self.im1=im1
        self.im2=im2
        self.im3=im3
        self.p1=p1
        self.p2=p2
        self.p3=p3
    def percent(self):
        ave1=self.im1/(self.im1+self.im2+self.im3)
        ave2=self.im2/(self.im1+self.im2+self.im3)
        ave3=self.im3/(self.im1+self.im2+self.im3)
        per=ave1*self.p1 + ave2*self.p2 + ave3*self.p3
        return "根据输入数据，预测此患者复发的概率为",per

#乳腺癌复发率与患者年龄相关性的图表
age=[25,35,45,55,65,75]   
p=relation(age)
print p.age()

#乳腺癌复发率与患者肿瘤大小相关性的图表
tumor=[2,7,12,17,22,27,32,37,42,47,52] 
q=relation(tumor)
print q.tumor()

#乳腺癌复发率与患者经历化疗情况的相关性的图表
ra=[0,1]
t=relation(ra)
print t.irradiat()

#乳腺癌复发率与患者肿瘤位置的相关性的图表
breast=[0,1]
b=relation(breast)
print b.breast()

#乳腺癌复发率与患者受侵淋巴结数的相关性的图表
inv_nodes=[1,4,7,10,13,16,25]
i=relation(inv_nodes)
print i.inv()

#乳腺癌复发率与患者恶性肿瘤程度相关性的图表
deg_malig=[1,2,3]
d=relation(deg_malig)
print d.deg()

#乳腺癌复发率与患者经历化疗情况的卡方分析
c=crosstab(fec.irradiat,fec.Class)
print c.tab()


#图形化界面设计
window=tk.Tk()

#窗口的标题
window.title('乳腺癌数据分析')

#窗口的大小
window.geometry('3000x3000') 

#添加标签label
l=tk.Label(window,text='预测乳腺癌是否复发',bg='green',font=('Arial,12'),width=30,height=3/2)    
l.pack()

m=tk.Label(window,text='请确认您的个人信息',bg='pink',font=('Arial,10'),width=30,height=1/2)    
m.pack()

#创建tkinter中的字符串变量（用作选择项的输入）    
var1=tk.StringVar()
var2=tk.StringVar()
var3=tk.StringVar()
var4=tk.StringVar()

def get_enter1():
    global m1 #把m1设成全局变量
    m1=var1.get()
    
def get_enter2():
    global m2 #把m2设成全局变量
    m2=var2.get()
    
def get_enter3():
    global m3 #把m3设成全局变量
    m3=var3.get()
    
def get_enter4():
    global m4 #把m4设成全局变量
    m4=var4.get()
    
def get_predict():
    #由前计算而来的相关性
    dic1={'20-29':0,'30-39':0.416667,'40-49':0.300000,'50-59':0.260417,'60-69':0.298246,'70-79':0.166667}
    dic2={'0-2':0.215962,'3-5':0.472222,'6-8':0.588235,'9-11':0.600000,'12-14':0.666667,'15-17':0.500000,'24-26':1}
    dic3={'1':0.169014,'2':0.215385,'3':0.529412}

    a1=dic1[m1]
    a2=0.164

    a3=dic2[m2]
    a4=0.677

    a5=dic3[m3]
    a6=0.919
    test=predict(a1,a2,a3,a4,a5,a6)
    result=test.percent()
    
    #t是输出文本框；insert指输出到鼠标光标当前位置；将result放入输出文本框中
    t.insert('insert',result)
    

    #列表里的元素是字符串
    
l1=tk.Label(window,bg='yellow',width=15,text='age')
l1.pack()

#text参数为显示在选择按钮上为20-29, 当选择第一个选择按钮的时候，把var赋值成value参数20-29
r11=tk.Radiobutton(window,text='20-29',variable=var1,value='20-29',command=get_enter1)
r11.pack()

r12=tk.Radiobutton(window,text='30-39',variable=var1,value='30-39',command=get_enter1)
r12.pack()

r13=tk.Radiobutton(window,text='40-49',variable=var1,value='40-49',command=get_enter1)
r13.pack()

r14=tk.Radiobutton(window,text='50-59',variable=var1,value='50-59',command=get_enter1)
r14.pack()

r15=tk.Radiobutton(window,text='60-69',variable=var1,value='60-69',command=get_enter1)
r15.pack()

r16=tk.Radiobutton(window,text='70-79',variable=var1,value='70-79',command=get_enter1)
r16.pack()





#第二个参数
l2=tk.Label(window,bg='yellow',width=15,text='inv-nodes')
l2.pack()

r21=tk.Radiobutton(window,text='0-2',variable=var2,value='0-2',command=get_enter2)
r21.pack()

r22=tk.Radiobutton(window,text='3-5',variable=var2,value='3-5',command=get_enter2)
r22.pack()

r23=tk.Radiobutton(window,text='6-8',variable=var2,value='6-8',command=get_enter2)
r23.pack()

r24=tk.Radiobutton(window,text='9-11',variable=var2,value='9-11',command=get_enter2)
r24.pack()

r25=tk.Radiobutton(window,text='12-14',variable=var2,value='12-14',command=get_enter2)
r25.pack()

r26=tk.Radiobutton(window,text='15-17',variable=var2,value='15-17',command=get_enter2)
r26.pack()

r27=tk.Radiobutton(window,text='24-26',variable=var2,value='24-26',command=get_enter2)
r27.pack()

#第三个参数
l3=tk.Label(window,bg='yellow',width=15,text='deg-mailg')
l3.pack()

r31=tk.Radiobutton(window,text='1',variable=var3,value='1',command=get_enter3)
r31.pack()

r32=tk.Radiobutton(window,text='2',variable=var3,value='2',command=get_enter3)
r32.pack()

r33=tk.Radiobutton(window,text='3',variable=var3,value='3',command=get_enter3)
r33.pack()

#第四个参数
l4=tk.Label(window,text='Please enter your tumor-size',bg='yellow',font=('Arial,10'),width=30,height=3/2)
l4.pack()
var4=tk.Entry(window,show=None) 
var4.pack()


#定义button，点击start to predict进行分析
b=tk.Button(window,text="Start to predict",width=20,height=2,command=get_predict)   
b.pack()  


#定义Text（输出的文本框）
t=tk.Text(window,height=2)
t.pack()

window.mainloop()



```
